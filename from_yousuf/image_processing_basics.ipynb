```python
import numpy as np 
```

```python
arr = np.random.randint(0, 256, size=(10, 10))
print(arr)
```

```python
import matplotlib.pyplot as plt
plt.imshow(arr, cmap='gray')
plt.axis('off')
plt.show()
```

To display an image from a NumPy array using Matplotlib, the imshow() function from matplotlib.pyplot is used.

```python
inverted_arr = 255 - arr
plt.imshow(inverted_arr, cmap='gray')
plt.axis('off')
plt.show()
```

```python
threshold = 128
binary_arr = np.where(arr > threshold, 255, 0)
plt.imshow(binary_arr, cmap='gray') 
plt.axis('off')
plt.show()
```

```python
croped_arr = arr[2:6, 2:6]
plt.imshow(croped_arr, cmap='gray')
plt.axis('off')     
plt.show()
```

```python
arr_edge = np.zeros_like(arr)
arr_edge[:, :-1] = arr[:, 1:] - arr[:, :-1]
plt.imshow(arr_edge, cmap='gray')
plt.axis('off')
plt.show()
```
