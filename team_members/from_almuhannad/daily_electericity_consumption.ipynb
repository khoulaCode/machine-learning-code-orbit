```python
import numpy as np
import pandas as pd

PATH = r"..\..\data\household_power_consumption\household_power_consumption.txt"
df = pd.read_csv(PATH, sep = ";")
df
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df["Date"])
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df["Datetime"] = pd.to_datetime(df["Date"], ' ',df['Time'], format='%d/%m/%Y %H:%M:%S')
df = df.set_index('Datetime')
```

```python
column_group = ['Global_reactive_power', 'Global_active_power','Voltage','Global_intensity', 'Sub_metering_1', 'Sub_metering_2',
       'Sub_metering_3']
df[column_group]= df[column_group].replace('?',np.nan)
```

```python
df[column_group]= df[column_group].astype('float64')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Global_active_power'], label='Global Active Power', color='blue')
plt.plot(df['Date'], df['Global_reactive_power'], label='Global Reactive Power', color='red')
plt.xlabel('Datetime')
plt.ylabel('Power (kW)')
plt.title('Global Active and Reactive Power Over Time')
plt.show()
```
