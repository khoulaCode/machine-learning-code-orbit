```python
import matplotlib.pyplot as plt
```

```python
path = r"C:\Users\admin\Downloads\archive(1)\Global_Economy_Indicators.csv"
df = pd.read_csv(path)
df
```

```python
df.columns
```

```python
#Top 5 countries by average GDP
top_5gdp = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending=False).head(5)
top_5gdp
```

```python
print(df.columns.tolist())
```

```python
#Countries with the Household consumption expenditure in the most recent year
recent_year = df[' Year '].max()
data = df[df[' Year '] == recent_year]
h_h_consumption = data[[' Country ',' Household consumption expenditure (including Non-profit institutions serving households) ']].sort_values(by=' Household consumption expenditure (including Non-profit institutions serving households) ', ascending=False).head(5)
h_h_consumption
```

```python
#Mean GDP per capita
gdp_per_capita = df.groupby(' Country ')[[" Gross Domestic Product (GDP) "," Per capita GNI "]].mean()
gdp_per_capita
```

```python
#Max imports of goods value
max_imp_value = df.groupby(' Country ')[' Imports of goods and services '].max().sort_values(ascending=False).head(5)
max_imp_value
```

```python
#Total exports over all years.
total_exports = df.groupby(' Country ', as_index=False)[' Exports of goods and services '].sum()
top_exports = total_exports.sort_values(by=' Exports of goods and services ', ascending=False).head(10)
top_exports
```

```python
#plot
Country_sum = df.groupby(' Country ').agg({
    ' Gross Domestic Product (GDP) ': "mean",
    ' Imports of goods and services ': "max",
    ' Exports of goods and services ': "sum"
}).rename(columns={
    ' Gross Domestic Product (GDP) ': "Mean GDP per Capita",
    ' Imports of goods and services ': "Max Import Value",
    ' Exports of goods and services ': "Total Exports"
})
Country_sum
```

```python
top_5gdp.plot(figsize=(6,5), color ='red', linestyle = '--')
plt.xlabel('Average GDP')
plt.title("Top 5 countries by average GDP")
plt.show
```

```python
#plot countries with the Household consumption expenditure in the most 
x = h_h_consumption[' Country ']
y = h_h_consumption[' Household consumption expenditure (including Non-profit institutions serving households) ']
plt.figure(figsize=(6,5))
plt.bar(x,y,color = 'powderblue', edgecolor= 'skyblue')
plt.title("Top Countries by Household Consumption Expenditure", fontweight='bold')
plt.xlabel('Country',fontsize = 10)
plt.ylabel('Household Consumption expenditure',fontsize = 10)
plt.grid(True, linestyle = '--')
```

```python
#plot Mean GDP per capita
top = gdp_per_capita.head(5)

plt.figure(figsize=(6,5))
plt.bar(top.index,top[' Gross Domestic Product (GDP) '],color='wheat',edgecolor='burlywood')
plt.title("Top Countries by Mean GDP per Capita", fontweight='bold')
plt.xlabel('Country',fontsize = 10)
plt.ylabel('Mean GDP per Capita',fontsize = 10)
plt.grid(True, linestyle = '--')
```

```python
#plot Max import of goods values

plt.figure(figsize=(6,5))
plt.bar(max_imp_value.index, max_imp_value.values, color='orangered')
plt.title("Top Countries by Max import of goods values", fontweight='bold')
plt.xlabel('Country',fontsize = 10)
plt.ylabel('Max import of goods values',fontsize = 10)
plt.grid(True, linestyle = '--')
```

```python
#Total exports over all years


x = top_exports[' Country ']
y = top_exports[' Exports of goods and services ']

plt.figure(figsize=(15,5))
plt.bar(x,y, color='blue')
plt.title("Total exports over all years", fontweight='bold')
plt.xlabel('Country',fontsize = 10)
plt.ylabel('Exports of goods and services',fontsize = 10)
plt.grid(True, linestyle = '--')
```
