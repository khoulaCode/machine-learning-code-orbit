```python
import pandas as pd
```

```python
train_path = r"C:\Users\admin\Downloads\walmart-recruiting-store-sales-forecasting\train.csv\train.csv"
train_df = pd.read_csv(train_path)
train_df
```

```python
features_path = r"C:\Users\admin\Downloads\walmart-recruiting-store-sales-forecasting\features.csv\features.csv"
features_df = pd.read_csv(features_path)
features_df
```

```python
stores_path = r"C:\Users\admin\Downloads\walmart-recruiting-store-sales-forecasting\stores.csv"
stores_df = pd.read_csv(stores_path)
stores_df
```

```python
merge_dfs = pd.merge(train_df,features_df,on = ['Store','Date'], how='left')

# Merge with stores 
all_dfs = pd.merge(merge_dfs,stores_df,on='Store', how='left')
all_dfs
```

```python
#use groupby
sales_by_store = all_dfs.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False)
print(sales_by_store)
```

```python
#onvert (Date) to datetime 
all_dfs['Date'] = pd.to_datetime(all_dfs['Date'])

#extract month column
all_dfs['Month'] = all_dfs['Date'].dt.month
```

```python
#Concatenating data
march_data = all_dfs[all_dfs['Month'] == 3]
may_data = all_dfs[all_dfs['Month'] == 5]

march_may_data = pd.concat([march_data, may_data], axis=0)
print(march_may_data)
```

```python
features_df = features_df.drop(columns=['IsHoliday'])
#use join
features_idx = features_df.set_index(['Store','Date'])
train_idx = train_df.set_index(['Store','Date'])

joined_data = train_idx.join(features_idx, how='left').reset_index()
joined_data
```
