```python
import pandas as pd 
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
path = r"C:\Users\admin\Downloads\archive (3)\household_power_consumption.txt"
df = pd.read_csv(path, sep=';')
df
```

```python
#Date and time set
df['Date_time'] = pd.to_datetime(df['Date'] +' '+ df['Time'], format='%d/%m/%Y %H:%M:%S')
df.set_index('Date_time', inplace = True)
df
```

```python
df.drop(columns=['Date','Time'], inplace=True)
df
```

```python
#slicing 
january2010 = df.loc['2010-01']
print(january2010)
```

```python
#resampling
#monthly avg consumption
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')
montly_avg_consm = df['Global_active_power'].resample('M').mean()
montly_avg_consm
```

```python
weekly_avg_consm = df['Global_active_power'].resample('W').mean()
weekly_avg_consm
```

```python
daily_avg_consm = df['Global_active_power'].resample('D').mean()
daily_avg_consm
```

```python
#Plotting 
plt.figure(figsize=(6,6))
plt.plot(montly_avg_consm, color = 'green')
plt.title('Monthly AVG Global Active Power Consumption')
plt.xlabel('Date')
plt.ylabel('Global Active Power')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(weekly_avg_consm, color = 'yellow')
plt.title('Weekly AVG Global Active Power Consumption')
plt.xlabel('Date')
plt.ylabel('Global Active Power')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(9,9))
plt.plot(daily_avg_consm, color = 'maroon')
plt.title('Daily AVG Global Active Power Consumption')
plt.xlabel('Date')
plt.ylabel('Global Active Power')
plt.grid(True)
plt.show()
```
