```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = r"..\data\Global_Economy_Indicators.csv"
df = pd.read_csv(path)
df
```

Top 5 countries by average GDP.

```python
gdp = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending=False).head(5)
gdp
```

Mean GDP per capita.

```python
gdp_per_capita = df.groupby(' Country ')[[" Gross Domestic Product (GDP) "," Per capita GNI "]].mean()
gdp_per_capita
```

Total exports over all years.

```python
total_exports_by_country = df.groupby([' Country ', ' Year '])[' Exports of goods and services '].sum()
total_exports_by_country
```

```python
plt.figure(figsize=(8, 5))
gdp.plot(kind='bar')
plt.title('Top 5 Countries by Average GDP')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
gdp_per_capita[' Gross Domestic Product (GDP) '].plot()
plt.title('Average GDP and Per Capita GNI by Country')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
total_exports_by_country.unstack().plot()
plt.title('Average GDP and Per Capita GNI by Country')
plt.show()
```
