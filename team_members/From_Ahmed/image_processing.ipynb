```python
import numpy as np
```

```python
Picture = np.random.randint(0,256,(10,10))
Picture
```

```python
import matplotlib.pyplot as plt
plt.imshow(Picture, cmap='gray')
plt.axis('off')
plt.show()
```

Matplotlib is a powerful and widely-used Python library for creating static, animated, and interactive visualizations

```python
inverted_Picture = 255 - Picture
plt.imshow(inverted_Picture, cmap='gray')
plt.axis('off')
plt.show()
```

```python
threshold = 128
thresholded_Picture = (Picture > threshold)*255  
plt.imshow(thresholded_Picture, cmap='gray')
plt.axis('off')
plt.show()
```

```python
cropped_Picture = Picture[2:6, 2:6]
plt.imshow(cropped_Picture, cmap='gray')
plt.axis('off')
plt.show()
```

```python
horizontal_edges = np.zeros_like(Picture)
horizontal_edges[:, :-1] = Picture[:, 1:] - Picture[:, :-1]
plt.imshow(horizontal_edges, cmap='gray')
plt.axis('off')
plt.show()
```
