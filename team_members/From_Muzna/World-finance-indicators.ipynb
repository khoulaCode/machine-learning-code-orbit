# Top 5 countries by average GDP.

```python
top5_gdp= df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending=False).head(5)
print(top5_gdp)
```

# Group by Country and calculate:

```python
# Mean GDP per capita.
mean_gdp=df.groupby(' Country ')[' Per capita GNI '].mean()
mean_gdp
```

```python
# Max unemployment rate.
# i did not fined any colum related to this 
```

```python
# Total exports over all years
total_exports=df.groupby(' Country ')[' Exports of goods and services '].sum()
total_exports
```

# plot

```python
import matplotlib.pyplot as plt
```

```python
# Plot Top 5 GDP
plt.figure(figsize=(10, 6))
top5_gdp.plot(kind='bar')
plt.title("Top 5 Countries by Average GDP")
plt.ylabel("GDP (USD)")
plt.show()
```

```python
top10 = mean_gdp.head(10)
plt.figure(figsize=(10, 6))
plt.bar(top10.index, top10.values, color='green')
plt.title("Top 10 Countries by Mean GDP Per Capita")
plt.xlabel("Country")
plt.ylabel("Mean Per Capita GNI (USD)")

plt.show()
```

```python
top10_exports = total_exports.head(10)
plt.figure(figsize=(10, 6))
plt.bar(top10_exports.index, top10_exports.values, color='royalblue')
plt.title("Top 10 Countries by Total Exports")
plt.xlabel("Country")
plt.ylabel("Total Exports (USD)")

plt.show()
```
