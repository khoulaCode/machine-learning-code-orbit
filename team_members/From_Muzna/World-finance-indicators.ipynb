# Group by Country and calculate:

```python
# Mean GDP per capita.
mean_gdp=df.groupby(' Country ')[' Per capita GNI '].mean()
mean_gdp
```

```python
top10 = mean_gdp.head(10)
plt.figure(figsize=(10, 6))
plt.bar(top10.index, top10.values, color='green')
plt.title("Top 10 Countries by Mean GDP Per Capita")
plt.xlabel("Country")
plt.ylabel("Mean Per Capita GNI")
plt.show()
```

```python
#Max import of goods values
max_exp=df.groupby(' Country ')[' Exports of goods and services '].max().sort_values(ascending=False)
max_exp
```

```python
top10 = max_exp.head(10)
plt.bar(top10.index, top10.values)
plt.xlabel("Country")
plt.show()
```

```python
# Total exports over all years
total_exports=df.groupby(' Country ')[' Exports of goods and services '].sum()
total_exports
```

```python
top10 = total_exports.head(20)
plt.figure(figsize=(25, 7))
plt.bar(top10.index, top10.values)
plt.xlabel("Country")
plt.show()
```
