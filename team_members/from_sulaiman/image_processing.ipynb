# Image Processing Basics (Grayscale  Conversion and Filtering)

## 1. Image Representation:

```python
import numpy as np

image_2d_array = np.random.randint(0, 256, (10, 10), dtype=np.uint8)
print(image_2d_array)
```

## 2. Display (Conceptual):

### I would use matplot and use the imshow function in matplot to tunr the array into an image, then show the image using the show function

```python
import matplotlib.pyplot as plt

# Display the image using matplotlib
plt.imshow(image_2d_array, cmap='gray', vmin=0, vmax=255)
plt.show()
```

## 3. Grayscale Inversion:

```python
inverted_image = 255 - image_2d_array
plt.imshow(inverted_image, cmap='gray', vmin=0, vmax=255)
plt.show()
```

## 4. Thresholding:

```python
threshold_value = 128
thresholded_image = np.where(image_2d_array < threshold_value, 0, 255)
plt.imshow(thresholded_image, cmap='gray', vmin=0, vmax=255)
plt.show()
```

## 5. Cropping:

```python
plt.imshow(thresholded_image[4:7, 4:7], cmap='gray', vmin=0, vmax=255)
plt.show()
```

## 6. Edge Detection (Simple - Convolution Concept):

```python
filtered_image = np.zeros_like(image_2d_array)
for i in range(len(image_2d_array)):
    for j in range(len(image_2d_array[i])):
        if j < len(image_2d_array[i]) - 1:
            # Simple edge detection by subtracting adjacent pixels
            temp = 0
            if image_2d_array[i, j+1] < image_2d_array[i, j]:
                temp = 255 + (int(image_2d_array[i, j+1]) - int(image_2d_array[i, j]))
            else:
                temp = (int(image_2d_array[i, j+1]) - int(image_2d_array[i, j]))
            filtered_image[i, j] = np.clip(temp, 0, 255).astype(np.uint8)
        else:
            filtered_image[i, j] = image_2d_array[i, j]
plt.imshow(filtered_image, cmap='gray', vmin=0, vmax=255)
```
