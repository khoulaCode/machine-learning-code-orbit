# World Finance Indicators

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
path = '../../data/Global_Economy_Indicators.csv'
df_GEI = pd.read_csv(path)
df_GEI
```

## Basic insights

### Getting the Top 5 countries by average GDP

```python
df_GEI_grouped = df_GEI.groupby([' Country '])
top_5_countries_average_GDP = df_GEI_grouped[[' Gross Domestic Product (GDP) ']].mean().sort_values(by=' Gross Domestic Product (GDP) ', ascending=False).head(5)
top_5_countries_average_GDP
```

### Countries with the Household consumption expenditure in the most recent year.

```python
df_GEI_filter_by_year = df_GEI[df_GEI[' Year '] == df_GEI[' Year '].max()]
household_consumption_txt = ' Household consumption expenditure (including Non-profit institutions serving households) '
countries_with_household_expense_filtered_year = df_GEI_filter_by_year[[' Country ', household_consumption_txt]].dropna()
countries_with_household_expense_filtered_year
```

## Group by Country and calculating

### Mean GDP per capita

```python
df_GEI_grouped[[' Gross Domestic Product (GDP) ']].mean()
```

### Max import of goods values

```python
df_GEI_grouped[[' Imports of goods and services ']].max()
```

### Total exports over all years

```python
df_GEI_grouped[[' Exports of goods and services ']].sum()
```

## Plots

### Plotting the Top 5 countries by average GDP

```python


plt.figure(figsize=(10, 6))
plt.bar(top_5_countries_average_GDP.index, top_5_countries_average_GDP[' Gross Domestic Product (GDP) '], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Gross Domestic Product (GDP)')
plt.title('Average GDP by Country')
plt.show()
```

### Countries with the Household consumption expenditure in the most recent year.

```python
top_5_countries_with_household_expense_filtered_year = countries_with_household_expense_filtered_year.sort_values(by=household_consumption_txt, ascending=False).head(5)
plt.bar(top_5_countries_with_household_expense_filtered_year[' Country '], top_5_countries_with_household_expense_filtered_year[household_consumption_txt], color='lightgreen')
plt.xlabel('Country')
plt.ylabel('Household Consumption Expenditure')
plt.title('Top 5 2021 Household Consumption Expenditure by Country')
plt.show()
```

### Plotting Mean GDP per capita

```python
plt.figure(figsize=(10, 6))
top_5_countries_average_GDP[' Gross Domestic Product (GDP) '].plot(kind='bar', color='skyblue')
plt.bar(top_5_countries_average_GDP.index, top_5_countries_average_GDP[' Gross Domestic Product (GDP) '], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Mean GDP')
plt.title('Top 5 Mean GDP per Country')
plt.show()
```

### Plotting max import of goods values

```python
top_countires_import_goods = df_GEI_grouped[[' Imports of goods and services ']].max().sort_values(by=' Imports of goods and services ', ascending=False).head(8)

plt.figure(figsize=(10, 6))
plt.bar(top_countires_import_goods.index, top_countires_import_goods[' Imports of goods and services '], color='orange')
plt.xlabel('Country')
plt.ylabel('Imports of Goods and Services')
plt.title('Top 5 Countries by Imports of Goods and Services')
plt.show()
```

### Plotting total exports over all years

```python
top_countries_total_exports = df_GEI_grouped[[' Exports of goods and services ']].sum().sort_values(by = ' Exports of goods and services ', ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_countries_total_exports.index, top_countries_total_exports[' Exports of goods and services '], color='blue')
plt.xlabel('Country')
plt.ylabel('Total Exports of Goods and Services')
plt.title('Top 5 Countries by Total Exports of Goods and Services')
plt.show()
```
