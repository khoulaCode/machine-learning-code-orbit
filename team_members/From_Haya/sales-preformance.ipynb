# Walmart Sales Complete Analysis
This notebook presents a full exploration and analysis of Walmart's sales data using pandas techniques.

```python
import pandas as pd
```

```python
train_df1 =r"C:\Users\admin\Desktop\data\train.csv"
train_df = pd.read_csv(train_df1 , low_memory=False)

features_df1 =r"C:\Users\admin\Desktop\data\features.csv"
features_df = pd.read_csv(features_df1 , low_memory=False)


stores_df1 =r"C:\Users\admin\Desktop\data\stores.csv"
stores_df= pd.read_csv( stores_df1, low_memory=False)



stores_df.head()
```

```python
features_df.head()
```

```python
train_df.head()
```

```python
# Merge the datasets
merged_df = pd.merge(train_df, features_df, on=["Store", "Date"], how="left")
final_df = pd.merge(merged_df, stores_df, on="Store", how="left")
```

```python

# Total sales by store and department
grouped_sales_by_store = final_df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False)
grouped_sales_by_dept = final_df.groupby('Dept')['Weekly_Sales'].sum().sort_values(ascending=False)
print("Top 5 stores by total sales:")
print(grouped_sales_by_store.head())
print("Top 5 departments by total sales:")
print(grouped_sales_by_dept.head())
```

```python

# Example of concatenation
df_part1 = final_df.head(5)
df_part2 = final_df.tail(5)
concat_df = pd.concat([df_part1, df_part2])
print("Concatenated result:")
print(concat_df)
```

```python
# EDA
print("Describe:")
print(final_df.describe())

print("\nValue Counts (Store):")
print(final_df["Store"].value_counts().head())

print("\nFiltering: Show only Type A stores:")
print(final_df[final_df["Type"] == "A"].head())
```

```python
# CONCAT: Concatenating Week 1 and Week 2 data
week1 = final_df[final_df['Date'] == "2010-02-05"]
week2 = final_df[final_df['Date'] == "2010-02-12"]
concatenated_weeks = pd.concat([week1, week2])
print("Concatenated two weeks of data:")
print(concatenated_weeks.head())
```
