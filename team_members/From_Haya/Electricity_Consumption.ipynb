# Electricity Consumption Time Series Analysis
This notebook demonstrates how to analyze household electricity consumption data using pandas time series capabilities.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

path=r"C:\Users\admin\Desktop\data\household_power_consumption.txt"
df= pd.read_csv(path, sep = ';', low_memory=False)
df
```

## Convert Date and Time to Datetime Index

```python
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S')
df.set_index('Datetime', inplace=True)
#df.drop(columns=['Date', 'Time'], inplace=True)
```

```python
df
```

## Time Slicing

```python
# Data for January 2007
january_2007 = df.loc['2007-01']
january_2007.head()
```

```python
# Data for a specific day
day_data = df.loc['2007-01-15']
day_data.head()
```

## Resampling Data

```python
df['Global_active_power']
```

```python
# Weekly average consumption
weekly = df['Global_active_power'].resample('W').mean()
weekly.head()
```

```python
# Monthly average consumption
monthly = df['Global_active_power'].resample('M').mean()
monthly.head()
```

## Plotting Monthly Consumption Trend

```python

monthly.plot(title='Monthly Global Active Power')
plt.xlabel('Month')
plt.ylabel('Power (kilowatts)')
plt.grid(True)
plt.show()
```

```python
weekly.plot(title='weekly Global Active Power')
plt.xlabel('Month')
plt.ylabel('Power (kilowatts)')
plt.grid(True)
plt.show()
```
