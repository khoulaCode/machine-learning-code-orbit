```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Global Economy Indicators.csv")
df.columns = df.columns.str.strip()
df
```

```python
# Top 5 countries

top_gdp = df.groupby('Country')['Gross Domestic Product (GDP)'].mean().sort_values(ascending=False).head(5)
print("Top 5 Countries by Average GDP: \n",top_gdp )
# plot
top_gdp.plot(kind='bar', title='Top 5 Countries by Average GDP', ylabel='Average GDP (in USD)', figsize=(10, 5), color='skyblue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Household consumption expenditure
most_recent_year = df['Year'].max()
column_name = 'Household consumption expenditure (including Non-profit institutions serving households)'
recent_data = df[df['Year'] == most_recent_year]
Household = recent_data[['Country', column_name]].sort_values(by=column_name, ascending=False).head(10)

print("Household Consumption Expenditure ", most_recent_year)
print(Household)

# Plot
Household.set_index('Country').plot(kind='bar', title='Household Consumption Expenditure' + str(most_recent_year), 
 ylabel='Expenditure (USD)',figsize=(10, 5), color='orange')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Mean GDP per capita
grouped = df.groupby('Country')
mean_gdp_per_capita = grouped['Per capita GNI'].mean().sort_values(ascending=False)
# plot
mean_gdp_per_capita.head(10).plot(kind='bar', title='Top 10 Countries by Mean GDP per Capita', ylabel='Mean GDP per Capita (GNI)', figsize=(10, 5), color='mediumseagreen')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Max import of goods values
import_col = 'Imports of goods and services'

max_imports = df.groupby('Country')[import_col].max().sort_values(ascending=False)

print("Top 10 Countries by Max Imports of Goods and Services:")
print(max_imports.head(10))

# Plot
max_imports.head(10).plot(kind='bar', title='Top 10 Countries by Max Imports of Goods and Services', ylabel='Max Import Value (USD)', figsize=(10, 5),
                          color='green')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Total exports over all years.
grouped = df.groupby('Country')
total_exports = grouped['Exports of goods and services'].sum().sort_values(ascending=False)
# plot
total_exports.head(10).plot(kind='bar', title='Top 10 Countries by Total Exports (1970–2021)', ylabel='Total Exports (USD)', figsize=(10, 5), color='steelblue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
