```python
## 1. Image Representation:

import numpy as np

image = np.random.randint(0, 256, (10, 10))
print("Original Image:\n", image)
```

```python
## 2. Display (Conceptual):

import matplotlib.pyplot as plt
plt.imshow(image, cmap='gray')
plt.show()
```

```python
## 3. Grayscale Inversion:

inverted_image = 255 - image
print("Inverted Image:\n", inverted_image)
```

```python
## 4. Thresholding:

thresholded_image = (image >= 128) * 255
print("Thresholded Image:\n", thresholded_image)
```

```python
## 5. Cropping:

cropped_image = image[3:7, 3:7]
print("Cropped 4x4 Center:\n", cropped_image)
```

```python
## 6. Edge Detection (Simple - Convolution Concept):

edge_image = np.zeros_like(image)
for i in range(10):
    for j in range(9):
        edge_image[i][j] = image[i][j+1] - image[i][j]

print("Edge Detection:\n", edge_image)
```
