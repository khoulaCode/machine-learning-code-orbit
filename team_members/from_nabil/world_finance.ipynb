```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = r"..\..\data\Global_Economy_Indicators.csv"
df = pd.read_csv(path)
df
```

Top 5 countries by average GDP.

```python
GDP_AVG = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending=False)
print("average GDP:", GDP_AVG.head())
```

Countries with the Household consumption expenditure in the most recent year.

```python
df.columns
```

```python
year_filter = df[df[' Year '] == df[' Year '].max()]
year_filter[[' Country ', ' Household consumption expenditure (including Non-profit institutions serving households) ']].dropna()
year_filter = year_filter.sort_values(by=' Household consumption expenditure (including Non-profit institutions serving households) ', ascending=False)
year_filter
```

Mean GDP per capita

```python
GDP_Per_capita = df.groupby(' Country ')[[' Gross Domestic Product (GDP) ',' Per capita GNI ']].mean()
GDP_Per_capita
```

Max import of goods values

```python
max_import = df.groupby(' Country ')[' Imports of goods and services '].max().sort_values(ascending=False)
max_import
```

Total exports over all years

```python
total_exports = df.groupby([' Country ', ' Year '])[' Imports of goods and services '].sum()
total_exports
```

Plot all above

```python
plt.figure(figsize=(10, 4))
GDP_AVG.head().plot(kind='bar')
plt.title('Top 5 Countries by Average GDP')
plt.show()
```

```python
plt.figure(figsize= (10,7))
GDP_Per_capita.plot(kind='line')
plt.title("mean of GDP per capita")
plt.xlabel('Country')
plt.ylabel('mean of GDP')
plt.show()
```

```python
filtered = total_exports.reset_index()
filtered = filtered[(filtered[' Year '] >= 2004) & (filtered[' Year '] <= 2021)]
filtered = filtered.set_index([' Country ', ' Year '])[' Imports of goods and services ']

plt.figure(figsize=(10,6))
filtered.unstack().plot()
plt.title('total exports over all years')
plt.xlabel('')
plt.ylabel('Exports')
plt.show()
```

```python
plt.figure(figsize=(6,10))
year_filter = year_filter.head()
year_filter.plot(kind='bar', x = ' Country ', y = ' Household consumption expenditure (including Non-profit institutions serving households) ')
plt.ylabel("Household consumption expenditure")
plt.title('Household consumption')
plt.legend().remove()
plt.show()
```

```python
plt.figure(figsize=(10,5))
max_import.plot(kind='line', color= 'black')
plt.title('Max import of goods values')
plt.show()
```
