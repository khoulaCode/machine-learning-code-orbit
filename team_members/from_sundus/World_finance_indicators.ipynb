```python
import seaborn as sns
```

```python
path = "Global Economy Indicators.csv"
df = pd.read_csv(path)
df
```

```python
most_recent_year = df['Year'].max()
print("Most Recent Year:", most_recent_year)
```

```python
top5_avg_gdp = df.groupby('Country')['Gross Domestic Product (GDP)'].mean().sort_values(ascending=False).head(5)
print("Top 5 countries by average GDP:\n",top5_avg_gdp)
```

```python
recent_data = df[df['Year'] == most_recent_year] 
```

```python
household_exp = recent_data[['Country','Household consumption expenditure (including Non-profit institutions serving households)'
]].dropna() 
```

```python
household_exp = household_exp.sort_values(by='Household consumption expenditure (including Non-profit institutions serving households)',
    ascending=False
) 
household_exp.head(10)
```

```python
country_sum = df.groupby('Country').agg({
    'Per capita GNI': 'mean',
    'Imports of goods and services': 'max',
    'Exports of goods and services': 'sum'
}).reset_index() 

country_sum.head()
```

```python
#Top 5 Countries by Average GDP plot

top5_avg_gdp.plot(figsize=(10, 5), color='skyblue')
plt.xlabel("Average GDP (USD)")
plt.title("Top 5 Countries by Average GDP (1970–2021)")
plt.gca().invert_yaxis()
plt.show() 
```

```python
#Plot Countries with the Household consumption expenditure in the most recent year
plt.figure(figsize=(10, 6))
sns.barplot(data=household_exp.head(5), 
            x='Household consumption expenditure (including Non-profit institutions serving households)', 
            y='Country')
plt.title('Top 10 Countries by Household Consumption Expenditure (2021)')
plt.xlabel('Expenditure (USD)')
plt.ylabel('Country')
plt.tight_layout()
plt.show() 
```
