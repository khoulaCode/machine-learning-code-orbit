## Merging different datasets merge() / join()

```python
merged_data = pd.merge(train, features, on=['Store', 'Date'])
merged_data.head()
```

```python
merged_df = pd.merge(train, features, how='left', on=['Store', 'Date'])
merged_df.head()
```

## Concatenating data

```python
first_week = train[train['Date'] == train['Date'].min()]  # Get the first week of data
second_week = train[train['Date'] == train['Date'].sort_values().unique()[1]]  # Get the second week of data
concatenated_data = pd.concat([first_week, second_week], ignore_index=True) # Concatenate the two weeks of data
concatenated_data.head()
```
