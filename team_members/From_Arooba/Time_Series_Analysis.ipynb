## Convert 'Date' and 'Time' columns to a single datetime column

```python
# Convert 'Date' and 'Time' columns to a single datetime column
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S')

df.columns  # Display the columns to verify the new column
```

```python
df.head() # Display the first 5 rows to verify the new column (DateTime)
```

```python
# Set the new datetime column as the index
df.set_index('DateTime', inplace=True) 
print(df.index) 
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time 

from datetime import time # Importing time class from datetime module

# Filter rows where 'Time' is greater than or equal to 23:00:00
filtered_df = df[df['Time'] >= time(23, 0, 0)]  
filtered_df
```

## Resampling

```python
daily_avg = df['Global_active_power'].resample('D').mean()
weekly_sum = df['Global_active_power'].resample('W').sum()
monthly_avg = df['Global_active_power'].resample('M').median()

print("Daily Avg:\n", daily_avg)
print("Weekly Sum:\n", weekly_sum.head())
print("Monthly Avg:\n", monthly_avg.head())
```

## Plotting simple trends

```python
import matplotlib.pyplot as plt

daily_avg.plot(title='Daily Average Power Usage')
plt.ylabel('Global Active Power (kilowatts)')
plt.xlabel('Date')

plt.show()
```

```python
import matplotlib.pyplot as plt

df['Date'] = pd.to_datetime(df['Date'], errors='coerce') # Convert 'Date' to datetime
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')  # Convert 'Global_active_power' to numeric
df['Global_reactive_power'] = pd.to_numeric(df['Global_reactive_power'], errors='coerce')  

# Drop rows with NaN values in 'Date', 'Global_active_power', and 'Global_reactive_power'
df.dropna(subset=['Date', 'Global_active_power', 'Global_reactive_power'], inplace=True)

# Plotting Global Active and Reactive Power over time
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Global_active_power'], label='Global Active Power', color='blue')
plt.plot(df['Date'], df['Global_reactive_power'], label='Global Reactive Power', color='red')
plt.xlabel('Datetime')
plt.ylabel('Power (kW)')
plt.title('Global Active and Reactive Power Over Time')
plt.show()
```
