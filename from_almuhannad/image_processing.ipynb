```python
import numpy as np
```

```python
image = np.random.randint(0,256,(10,10))
image
```

```python
import matplotlib.pyplot as plt
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

</p1> matplotlib.pyplot.imshow() is used to display an array as an image. Since the image is in grayscale, we need to set the colormap to "gray". After that, we can show the image using plt.show(). <p1/>

```python
inverted_image = 255 - image
plt.imshow(inverted_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
threshold = 128
thresholded_image = (image > threshold).astype(np.uint8) * 255
plt.imshow(thresholded_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
cropped_image = image[2:6, 2:6]
plt.imshow(cropped_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
horizontal_edges = np.zeros_like(image)
horizontal_edges[:, :-1] = image[:, 1:] - image[:, :-1]
plt.imshow(horizontal_edges, cmap='gray')
plt.axis('off')
plt.show()
```
