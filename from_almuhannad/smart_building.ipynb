```python
import numpy as np
```

```python
time_in_minutes = np.arange(1440)
print(time_in_minutes)
```

```python
temp = np.random.normal(22,5,size=1440)
print(temp)
```

```python
humidity = np.random.normal(50,10,size=1440)
print(humidity)
```

```python
pressure = np.random.normal(1012,10,size=1440)
print(pressure)
```

```python
data = np.column_stack((time_in_minutes,temp,humidity,pressure))
data
```

```python
avg_temp = np.mean(temp)
print (avg_temp)
avg_hum = np.mean(humidity)
print (avg_hum)
avg_pressure = np.mean(pressure)
print (avg_pressure)
```

```python
min_temp = np.min(temp)
print (min_temp)
min_hum = np.min(humidity)
print (min_hum)
min_pressure = np.min(pressure)
print (min_pressure)
```

```python
max_temp = np.max(temp)
print (max_temp)
max_hum = np.max(humidity)
print (max_hum)
max_pressure = np.max(pressure)
print (max_pressure)
```

```python
std_temp = np.std(temp)
print (std_temp)
std_hum = np.std(humidity)
print (std_hum)
std_pressure = np.std(pressure)
print (std_pressure)
```

```python
for hour in range(24):
    hour_count = hour + 1
    start_index = hour * 60
    end_index = start_index + 60
    avg_hour_tempreture = np.mean(temp[start_index:end_index])
    print(f" {hour_count} : {avg_hour_tempreture}")
```

```python
for hour in range(24):
    hour_count = hour + 1
    start_index = hour * 60
    end_index = start_index + 60
    avg_hour_humidity = np.mean(humidity[start_index:end_index])
    print(f" {hour_count} : {avg_hour_humidity}")
```

```python
for hour in range(24):
    hour_count = hour + 1
    start_index = hour * 60
    end_index = start_index + 60
    avg_hour_pressure = np.mean(pressure[start_index:end_index])
    print(f" {hour_count} : {avg_hour_pressure}")
```

```python
temp_anomaly = temp < 25
for i in range(len(temp_anomaly)):
    if temp_anomaly[i]:
        print(f"minute {i}: {temp[i]}")
```

```python
min_temp = np.min(temp)
time_for_min_temp = np.where(temp == min_temp)[0][0]
print(f"minimum temp {min_temp}: at minute {time_for_min_temp}")
```
