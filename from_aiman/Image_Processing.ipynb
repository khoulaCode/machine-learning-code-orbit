```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Image Representation (Random 10x10)

```python
pic_matrix = np.random.randint(0, 256, size=(10, 10), dtype=np.uint8)
print("1. Random Grayscale Image:\n", pic_matrix)
```

```python
import matplotlib.pyplot as plt
plt.imshow(pic_matrix, cmap='gray')
plt.axis('off')
plt.show()
```

# 2. Display the Image (Concept)

```python
plt.imshow(pic_matrix, cmap='gray')
plt.colorbar()
plt.title("Grayscale Image Display")
plt.show()
```

# 3. Grayscale Inversion

```python
inverse_gray = 255 - pic_matrix
print("\n2. Inverted Image:\n", inverse_gray)
```

```python
import matplotlib.pyplot as plt
plt.imshow(inverse_gray, cmap='gray')
plt.axis('off')
plt.show()
```

# 4. Thresholding (Binary)

```python
cutoff = 128
binary_output = np.where(pic_matrix < cutoff, 0, 255)
print("\n3. Binary Image (after thresholding):\n", binary_output)
```

```python
import matplotlib.pyplot as plt
plt.imshow(binary_output, cmap='gray')
plt.axis('off')
plt.show()
```

# 5. Cropping (4x4 Center)

```python
row_start = (pic_matrix.shape[0] - 4) // 2
col_start = (pic_matrix.shape[1] - 4) // 2
roi_crop = pic_matrix[row_start:row_start + 4, col_start:col_start + 4]
print("\n4. Cropped 4x4 Center:\n", roi_crop)
```

```python
import matplotlib.pyplot as plt
plt.imshow(roi_crop, cmap='gray')
plt.axis('off')
plt.show()
```

# 6. Edge Detection (Basic Horizontal Convolution)

```python
horizontal_edges = np.zeros_like(pic_matrix, dtype=int)
for row in range(pic_matrix.shape[0]):
    for col in range(pic_matrix.shape[1] - 1):
        horizontal_edges[row, col] = int(pic_matrix[row, col + 1]) - int(pic_matrix[row, col])
print("\n5. Horizontal Edge Detection:\n", horizontal_edges)
```

```python
import matplotlib.pyplot as plt
plt.imshow(horizontal_edges, cmap='gray')
plt.axis('off')
plt.show()
```
