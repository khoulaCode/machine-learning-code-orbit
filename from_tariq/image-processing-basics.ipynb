```python
import numpy as np
```

```python
image = np.random.randint(0, 256, (10, 10))
print(image)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

</p1> matplotlib.pyplot.imshow() allows you to render an arry as an image, since it's gray scale we have to specify the colormap "gray" then we can display the image using plt.show() <p1/>

```python
inverted_image = 255 - image
plt.imshow(inverted_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
threshold = 128
thresholded_image = (image > threshold).astype(np.uint8) * 255
plt.imshow(thresholded_image, cmap='gray')
plt.axis('off')
plt.show()

```

```python
cropped_image = image[2:6, 2:6]
plt.imshow(cropped_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
horizontal_edges = np.zeros_like(image)
horizontal_edges[:, :-1] = image[:, 1:] - image[:, :-1]
plt.imshow(horizontal_edges, cmap='gray')
plt.axis('off')
plt.show()
```
