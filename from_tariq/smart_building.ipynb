```python
import numpy as np
```

```python
time_in_minutes = np.arange(1440)
print(time_in_minutes)
```

```python
temp = np.random.normal(22,5,size=1440)
temp = temp.astype(int)
print(temp)
```

```python
humidity = np.random.normal(50,10,size=1440)
humidity = humidity.astype(int)
print(humidity)
```

```python
pressure = np.random.normal(1012,10,size=1440)
pressure = pressure.astype(int)
print(pressure)
```

```python
data = np.column_stack((time_in_minutes,temp,humidity,pressure))
print(data)
```

```python
avg_temp = np.mean(temp)
print (f"Average Temperature: {avg_temp:.2f}°C")
avg_humidity = np.mean(humidity)
print (f"average Humidity: {avg_humidity:.2f}%")
avg_pressure = np.mean(pressure)
print (f"Average Pressure: {avg_pressure:.2f}hPa")
```

```python
min_temp = np.min(temp)
print(f"minimum Temperature: {min_temp}°C")
min_humidity = np.min(humidity)
print(f"minimum Humidity: {min_humidity}%")
min_pressure = np.min(pressure)
print(f"minimum Pressure: {min_pressure}hPa")
```

```python
max_temp = np.max(temp)
print(f"maximum Temperature: {max_temp}°C")
max_humidity = np.max(humidity)
print(f"maximum Humidity: {max_humidity}%")
max_pressure = np.max(pressure)
print(f"maximum Pressure: {max_pressure}hPa")
```

```python
std_temp = np.std(temp)
print(f"Standard Deviation of Temperature: {std_temp:.2f}°C")
std_humidity = np.std(humidity)
print(f"Standard Deviation of Humidity: {std_humidity:.2f}%")
std_pressure = np.std(pressure)
print(f"Standard Deviation of Pressure: {std_pressure:.2f}hPa")
```

```python
for hour in range(24):
    hour_count = hour + 1
    start_index = hour * 60
    end_index = start_index + 60
    avg_hour_tempreture = np.mean(temp[start_index:end_index])
    print(f"Average Temperature for hour {hour_count}: {avg_hour_tempreture:.2f}°C")
```

```python
for hour in range(24):
    hour_count = hour + 1
    start_index = hour * 60
    end_index = start_index + 60
    avg_hour_humidity = np.mean(humidity[start_index:end_index])
    print(f"Average Humidity for hour {hour_count}: {avg_hour_humidity:.2f}%")
```

```python
for hour in range(24):
    hour_count = hour + 1
    start_index = hour * 60
    end_index = start_index + 60
    avg_hour_pressure = np.mean(pressure[start_index:end_index])
    print(f"Average Pressure for hour {hour_count}: {avg_hour_pressure:.2f}hPa")
```

```python
temp_anomaly = temp < 25
for i in range(len(temp_anomaly)):
    if temp_anomaly[i]:
        print(f"Temperature anomaly detected at minute {i}: {temp[i]}°C")
```

```python
min_temp = np.min(temp)
time_for_min_temp = np.where(temp == min_temp)[0][0]
print(f"Minimum temperature of {min_temp}°C occurred at minute {time_for_min_temp}.")
```
