Using groupby() to group Columns by other Columns to perform aggregation functions and such

```python
temp_df_csv = df_csv[[' CountryID ', ' Year ', ' Population ']]
```

```python
print(temp_df_csv.groupby([' CountryID ', ' Year '])[' Population '].sum())
```

```python
print(temp_df_csv.groupby([' CountryID '])[' Population '].mean())
```

```python
print(temp_df_csv.groupby([' CountryID '])[' Population '].aggregate(['mean', 'sum', 'count']))
```

you can use more than one aggregation function for a column

```python
print(temp_df_csv.groupby([' CountryID '])[' Population '].agg(['mean', 'sum', 'count']))
```

merging 2 tables with each other using diferent methods

```python
df_customers = pd.DataFrame({
    'CustomerID' : [1, 2, 3, 4],
    'Name' : ['Alice', 'Bob', 'Charlie', 'David']
})
df_customers
```

```python
df_orders = pd.DataFrame({
    'OrderID' : [101, 102, 103, 104],
    'CustomerID' : [2, 4, 1, 5],
    'Amount' : [100, 250, 150, 300]
})
df_orders
```

```python
pd.merge(df_customers, df_orders, on='CustomerID', how='inner')
```

```python
pd.merge(df_customers, df_orders, on='CustomerID', how='outer')
```

```python
pd.merge(df_customers, df_orders, on='CustomerID', how='left')
```

```python
pd.merge(df_customers, df_orders, on='CustomerID', how='right')
```

## Serial Plotting

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
x
```

```python
y
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

```python
y2 = np.cos(x)
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot(x, y, label='sin(x)', color = 'red', linestyle = '-.')
plt.plot(x, y2, label='cos(x)', color = 'black', linestyle = ':')
plt.title('Sine and Cosine Waves', fontsize = 16)
plt.legend()
plt.show()
```

```python
zeros = np.zeros((10))
```

```python
plt.figure(figsize=(15,7))
plt.plot(x, y, label='sin(x)', color = 'red', linestyle = '-.')
plt.plot(x, y2, label='cos(x)', color = 'black', linestyle = ':')
plt.plot(zeros, label='zeros', color='black')
plt.title('Sine and Cosine Waves', fontsize = 16, fontweight='bold')
plt.xlabel('X-values', fontsize = 16, fontweight='bold')
plt.ylabel('Y-values', fontsize = 16, fontweight='bold')
plt.legend()
plt.grid(True)
plt.show()
```

## numerical plotting

```python
np.random.seed(42)
x_scatter = np.random.rand(50)*10
y_scatter = np.random.rand(50)*10+x_scatter*0.5

plt.figure(figsize = (15,15))
plt.scatter(x_scatter, y_scatter, color='red', alpha=0.7)
plt.title("Scatter Plot of X vs Y", fontsize=14)
plt.xlabel('X-values')
plt.ylabel('Y-values')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

## categorical plotting

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 44, 89, 13]

plt.figure(figsize = (12,9))
plt.bar(categories, values, color=['plum', 'springgreen', 'olivedrab', 'skyblue'])
plt.title("Bar Plot of Category vs Values", fontsize=14)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```
