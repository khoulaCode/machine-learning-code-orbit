1. Data Generation

```python
import numpy as np
time_in_minutes = np.arange(1440)
temperature = np.random.normal(22,5 ,size=1440)
humidity = np.random.normal(55,10 ,size=1440) 
pressure = np.random.normal(1012,10 ,size=1440)
data = np.column_stack((time_in_minutes, temperature, humidity, pressure))
data
```

2. Basic Statistics

A. average

```python
tem_avg = np.mean(temperature)
print("average of temperature:\n",tem_avg)
hum_avg = np.mean(humidity)
print("average of humidity:\n",hum_avg)
pre_avg = np.average(pressure)
print("average of pressure:\n",pre_avg)
```

B. maximum 

```python
tem_max = np.max(temperature)
print("maximum of temperature:\n",tem_max)
hum_max = np.max(humidity)
print("maximum of humidity:\n",hum_max)
pre_max = np.max(pressure)
print("maximum of pressure:\n",pre_max)
```

C. minimum

```python
tem_min = np.min(temperature)
print("minimum of temperature:\n", tem_min)
hum_min = np.min(humidity)
print("minimum of humidity:\n", hum_min)
pre_min = np.min(pressure)
print("minimum of pressure:\n", pre_min)
```

D. standard deviation

```python
tem_sd = np.std(temperature)
print("standard deviation of temperature: \n", tem_sd)
hum_sd = np.std(humidity)
print("standard deviation of humidity: \n", hum_sd)
pre_sd = np.std(pressure)
print("standard deviation of pressure: \n", pre_sd)
```

4. Hourly Averages

```python
for hour in range(24):
    each_hour = hour + 1
    str_ind = hour *60
    end_ind = str_ind + 60
    tem_avg_hour = np.mean(temperature[str_ind:end_ind])
    print(f"avrage temperature for hour {each_hour}: {tem_avg_hour}")
```

4. Anomaly Detection

```python
tem_ano = temperature < 25
for i in range(len(tem_ano)):
    if tem_ano[i]:
        print(f"Temperature anomaly detected {i}: {temperature[i]}")
```

```python
min_tem = np.min(temperature)
mini_tem = np.where(temperature == min_tem)[0][0]
print(f"Minimum temperature  {min_tem}  {mini_tem}.")
```
